# Video to Frames

Using OpenCV, extract 1 frame per second for every video.

```python
import cv2
import math
import os
import time
import matplotlib.pyplot as plt 
```

```python
start_time = time.time()

videos_dir = "../data/train/videos"

for videofile in os.listdir(videos_dir):
    if not os.path.exists("../data/train/frames/{}".format(videofile.split(".")[0])):
        os.makedirs("../data/train/frames/{}".format(videofile.split(".")[0]))
    if not videofile.split(".")[1] == 'mp4':
        continue
    else:
        count = 0
        print('Extracting frames from {}'.format(videofile))
        cap = cv2.VideoCapture(videofile)   # capturing the video from the given path
        frameRate = cap.get(5)              # frame rate
        x=1
        while(cap.isOpened()):
            frameId = cap.get(1)            # current frame number
            ret, frame = cap.read()
            if (ret != True):
                break
            if (frameId % math.floor(frameRate) == 0):
                filename ="../data/train/frames/{}/frame{}.png".format(videofile.split(".")[0], count)
                count+=1
                cv2.imwrite(filename, frame)
        cap.release()
        #cv2.destroyAllWindows()
    
print ("Done! .......... {} seconds".format(round(time.time()-start_time, 2)))
```
